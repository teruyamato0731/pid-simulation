```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axisartist.axislines import SubplotZero

# 数式はTexのフォントで表示
plt.rcParams['mathtext.fontset'] = 'cm'

# 操作量
U = 0.5
e = np.linspace(-0.75, 0.75, 100)
Kp = 1
# 操作量
u = Kp * e
u[u > U] = U
u[u < -U] = -U

fig = plt.figure()
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

ax.plot(0, 0, color='black', marker='o')

ax.plot(e, u, color='orange')

ax.plot(e[e > 0], U * np.ones(50), '--', color='red')
ax.scatter(0, U, facecolor='w', edgecolors='red', marker='o')
ax.text(-0.05, U, r'$u_{\mathrm{max}}$', fontsize=20, horizontalalignment='right')

ax.plot(e[e < 0], -U * np.ones(50), '--', color="blue")
ax.scatter(0, -U, facecolor='w', edgecolors='blue', marker='o')
ax.text(0.05, -U, r'$u_{\mathrm{min}}$', fontsize=20, horizontalalignment='left')

ax.text(0.75, -0.05, r"$e(t)$", fontsize=20, verticalalignment='top')
ax.text(0.05, 0.6, r"$u(t)$", fontsize=20, horizontalalignment='left')
ax.text(0.01, -0.01, r"$O$", fontsize=20, horizontalalignment='left', verticalalignment='top')

for direction in ["xzero", "yzero"]:
  # (0,0)を通る軸を表示する
  ax.axis[direction].set_visible(True)

  # 各軸の端を矢印にする
  ax.axis[direction].set_axisline_style("->", size=2)

for direction in ["left", "right", "bottom", "top"]:
  # 図の枠を非表示にする
  ax.axis[direction].set_visible(False)

ax.axis['left'].set_axisline_style("->")

ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
# ax.axes.xaxis.set_ticklabels([])
# ax.axes.yaxis.set_ticklabels([])

plt.show()
```
